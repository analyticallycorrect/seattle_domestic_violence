```python
myshp = open("../data/Seattle Police Micro-Community Policing Plans Neighborhoods/geo_export_f0b88624-4830-45b3-be26-feef64800898.shp", "rb")
mydbf = open("../data/Seattle Police Micro-Community Policing Plans Neighborhoods/geo_export_f0b88624-4830-45b3-be26-feef64800898.dbf", "rb")
```

```python
# read the shapefile
reader = shapefile.Reader(shp=myshp, dbf=mydbf)
fields = reader.fields[1:]
field_names = [field[0] for field in fields]
buffer = []
for sr in reader.shapeRecords():
   atr = dict(zip(field_names, sr.record))
   geom = sr.shape.__geo_interface__
   buffer.append(dict(type="Feature", \
    geometry=geom, properties=atr)) 

# write the GeoJSON file
from json import dumps
geojson = open("../data/neighborhood_shapes.json", "w")
geojson.write(dumps({"type": "FeatureCollection",\
"features": buffer}, indent=2) + "\n")
geojson.close()
```
